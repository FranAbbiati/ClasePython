## 1. Funciones (`def`)

En Python, eso es una función `def`.

```python
# Definimos la función (La enseñamos a Python)
def saludar(nombre):
    print("Hola", nombre)
    print("¡Espero que tengas un buen día!")
nombre=input("como te llamas :")
saludar(nombre)
```

### 1.1 `return` vs `print`

Esto es vital:
* `print`: Muestra algo en pantalla, pero el dato se pierde en el aire.
* `return`: La función te **devuelve** el dato para que lo guardes en una variable y lo sigas usando.

```python
# Función que SOLO muestra (print)
def sumar_pantalla(a, b):
    print(a + b)

# Función que DEVUELVE el valor (return)
def sumar_real(a, b):
    resultado = a + b
    return resultado

valor=sumar_pantalla(24,6)
print(valor+10)
```

### 4. Mayor de Tres

**Consigna:**
1. Crea una función `mayor_de_tres(n1, n2, n3)`.
2. Usa condicionales (`if/elif/else`) para ver cuál es el más grande.
3. Retorna el número mayor.

```python
# Tu código aquí 👇

def mayorDeTres(n1, n2, n3):
    if n1>n2 and n1>n3:
        return n1
    elif n2>n1 and n2>n3:
        return n2
    elif n3>n1 and n3>n2:
        return n3
    else:
        print("No hay un solo numero mayor")
    
n1T=input("Dame un numero: ")
n2T=input("Dame un numero: ")
n3T=input("Dame un numero: ")
n1=float(n1T)
n2=float(n2T)
n3=float(n3T)

print(mayorDeTres(n1, n2, n3))
```

### 6. Conversor de Minutos a Horas

**Consigna:**
1. Crea una función `minutos_a_horas(minutos)`.
2. Calcula cuántas horas y cuántos minutos sobran (usa división entera `//` y módulo `%`).
3. Imprime algo como: **"2 horas y 30 minutos"**.

```python
# Tu código aquí 👇

def min_horas(minutos):
    hours=minutos//60
    minutes=(minutos-(hours*60))
    print(f"{hours} horas y {minutes} minutos")

min=input("Decime los minutos: ")
minutos=int(min)

print(min_horas(minutos))
```

### 8. Calculadora de Descuento

**Consigna:**
1. Crea una función `calcular_precio_final(precio, descuento)`.
2. Calcula el precio con el descuento aplicado.
3. Retorna el precio final.

```python
# Tu código aquí 👇

def calcularPrecioFinal(precio, descuento):
    precioFinal=precio-descuento
    return precioFinal

precioT=input("Decime el precio: ")
precio=float(precioT)
descuentoT=input("Decime el descuento: ")
descuentoN=float(descuentoT)
descuento=descuentoN*100/precio

print(f"El precio final es {calcularPrecioFinal(precio,descuento)}")

```
